# A complete example

```python
x = np.arange(1, 5)
plt.plot(x, x*1.5, label='Normal')
plt.plot(x, x*3.0, label='Fast')
plt.plot(x, x/3.0, label='Slow')
plt.legend()
plt.grid(True)
plt.title('Sample Growth of a Measure')
plt.xlabel('Samples')
plt.ylabel('Values Measured')
plt.legend(loc='upper left')
```

# Save as a picture file

```python
x = np.arange(1, 5)
plt.plot(x, x*1.5, label='Normal')
plt.plot(x, x*3.0, label='Fast')
plt.plot(x, x/3.0, label='Slow')
plt.legend()
plt.grid(True)
plt.title('Sample Growth of a Measure')
plt.xlabel('Samples')
plt.ylabel('Values Measured')
plt.legend(loc='upper left')
# this is the code to save as a picture
plt.savefig('haha.png')
```

# change line color

| abbreviation | Color Name |
| ------------- |-----------|
| b |	blue |
|c | cyan |
|g | green |
|k | black |
|m | magenta |
|r | red |
|w | white |
|y | yellow |

```python
y = np.arange(1, 3)
plt.plot(y, 'y', y+1, 'm', y+2, 'c');
```

# line style

| abbreviation | style |
| ------------ | ----- |
| - | solid line |
| -- | dashed line |
| -. | dash-dot line |
| : | dotted line |

```python
y = np.arange(1, 3)
plt.plot(y, '--', y+1, '-.', y+2, ':')
```

# marker style

```python
y = np.arange(1, 3, 0.2)
plt.plot(y, 'x', y+0.5, 'o', y+1, 'D', y+1.5, '^', y+2, 's')
```

# line and marker

```python
y = np.arange(1, 3, 0.3)
plt.plot(y, 'cx--', y+1, 'mo:', y+2, 'kp-.');
```

# Finer control with keyword arguments

```python
y = np.arange(1, 3, 0.3)
plt.plot(y, 
         color='blue', linestyle='dashdot', 
         linewidth=4, marker='o', 
         markerfacecolor='red', markeredgecolor='black', 
         markeredgewidth=3, markersize=12)
```
